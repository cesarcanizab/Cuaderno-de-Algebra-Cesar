```python
#Ejercicios vectores ortogonales y ortonormales 
# a) Determine si los siguientes conjuntos de vectores es ortogonal, si es ortogonal determine si es también ortonormal y determine si es base para R3
#$1) V_1 = \{(2,-4),(2,1)\}$
#$2) V_2 = \{(3,2),(-4,-5)\}$
#$3) V_3 = \{(4,-1,1),(-1,0,4),(-4,-17,-1)\}$
```

```python
import numpy as np
#Opcion 1
# Definir los vectores
v1 = np.array([[2,-4],[2,1]])
v2 = np.array([[3,2],[-4,-5]])
v3 = np.array([[4,-1,1],[-1,0,4],[-4,-17,-1]])

# Comprobar si los vectores son ortogonales
def are_orthogonal(vectors):
    for i in range(len(vectors)):
        for j in range(i+1, len(vectors)):
            if np.dot(vectors[i], vectors[j]) != 0:
                return False
    return True

if are_orthogonal(v1):
    print("v1 es un conjunto de vectores ortogonales.")
    # Comprobar si los vectores son ortonormales
    if np.linalg.norm(v1[0]) == 1 and np.linalg.norm(v1[1]) == 1:
        print("v1 es un conjunto de vectores ortonormales.")
    else:
        print("v1 no es un conjunto de vectores ortonormales.")
    # Comprobar si el conjunto es una base para R3
    if np.linalg.matrix_rank(v1) == 3:
        print("v1 es una base para R3.")
    else:
        print("v1 no es una base para R3.")
else:
    print("v1 no es un conjunto de vectores ortogonales.")

if are_orthogonal(v2):
    print("v2 es un conjunto de vectores ortogonales.")
    # Comprobar si el conjunto es una base para R3
    if np.linalg.matrix_rank(v2) == 3:
        print("v2 es una base para R3.")
    else:
        print("v2 no es una base para R3.")
else:
    print("v2 no es un conjunto de vectores ortogonales.")

if are_orthogonal(v3):
    print("v3 es un conjunto de vectores ortogonales.")
    # Comprobar si los vectores son ortonormales
    if np.linalg.norm(v3[0]) == 1 and np.linalg.norm(v3[1]) == 1 and np.linalg.norm(v3[2]) == 1:
        print("v3 es un conjunto de vectores ortonormales.")
    else:
        print("v3 no es un conjunto de vectores ortonormales.")
    # Comprobar si el conjunto es una base para R3
    if np.linalg.matrix_rank(v3) == 3:
        print("v3 es una base para R3.")
    else:
        print("v3 no es una base para R3.")
else:
    print("v3 no es un conjunto de vectores ortogonales.")
```
